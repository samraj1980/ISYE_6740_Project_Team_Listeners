```python
# Create the combined topic columns 
topic_modelling["combined_Probability"] = topic_modelling["probable_Topic_NMF"] + '-' + topic_modelling["probable_Topic_LSA"] + '-' + topic_modelling["probable_Topic_LDA"]
#topic_modelling["combined_Probability"] = topic_modelling["probable_Topic_NMF"] + '-' + topic_modelling["probable_Topic_LSA"]
```

```python
#Creating the Dictionary column 

def create_dictionary(combined):
    d = dict()
    
    words = str(combined).split('-')
        
    for word in words: 
        # Check if the word is already in dictionary 
        if word in d: 
            # Increment count of word by 1 
            d[word] = d[word] + 1
        else: 
            # Add the word to dictionary with count 1 
            d[word] = 1
            
    return d


topic_modelling['dictionary'] = topic_modelling['combined_Probability'].apply (lambda x: create_dictionary(x))
```

```python
# Create a column with ensembled topic names 

def create_final_topic_name(combined):
    #final_topic = sorted(combined.keys(), key=lambda k: combined[k], reverse=True)
    #print(final_topic)
    matching = [k for k,v in combined.items() if float(v) >= 2]
    matching = '-'.join(matching)

    #if any(final_topic[0] in s for s in unique_topics):
    #    matching = [s for s in unique_topics if final_topic[0] in s]
    #else:
    #    print(final_topic[0])
    return(matching)

topic_modelling['final_topic'] = topic_modelling['dictionary'].apply (lambda x: create_final_topic_name(x))

#topic_modelling['final_topic'] = topic_modelling['final_topic'].replace('climate crisis', 'climate change')
#topic_modelling['final_topic'] = topic_modelling['final_topic'].replace('president impeachment', 'impeachment inquiry')
```

```python
# Load the topics across party lines and create the unique topic names 
import pickle

#Load the probable topic names 
with open('topic_names_rep.pkl', 'rb') as f:
        topic_names_rep = pickle.load(f)

with open('topic_names_dem.pkl', 'rb') as f:
        topic_names_dem = pickle.load(f)

unique_topics = list(set(topic_names_dem+topic_names_rep))
unique_topics.sort()
```

```python
# Create the Unique topic names 
del unique_topics[19] # remove mental health 
del unique_topics[16] # remove health crisis
del unique_topics[21] # remove postal service
del unique_topics[2] # remove article impeachment
del unique_topics[4] # remove Business Owner
unique_topics.append('vote election mail')
unique_topics.append('pandemic relief response')
unique_topics.append('job economy unemployment')
unique_topics.append('president impeachment')
unique_topics.append('covid19 testing')
unique_topics.append('trump administration')
unique_topics.append('farmer tradewar')
unique_topics.append('veteran service')
unique_topics.append('drug prescription')
unique_topics.append('legislation bipartisan')
unique_topics.append('relief payment')
unique_topics.sort()
```

```python
# Create a column with ensembled topic names 

def create_final_normalized_topic_name(combined):
    #final_topic = sorted(combined.keys(), key=lambda k: combined[k], reverse=True)
    final_topic = combined.split('-')
    
    if any(final_topic[0] in s for s in unique_topics):
        matching = [s for s in unique_topics if final_topic[0] in s]
    else:
        print(final_topic[0])
    return(matching[0])

topic_modelling_refined['final_normalized_topic'] = topic_modelling_refined['final_topic'].apply (lambda x: create_final_normalized_topic_name(x))

topic_modelling_refined['final_normalized_topic'] = topic_modelling_refined['final_normalized_topic'].replace('climate crisis', 'climate change')
topic_modelling_refined['final_normalized_topic'] = topic_modelling_refined['final_normalized_topic'].replace('president impeachment', 'impeachment inquiry')
```
